```python
plt.figure(figsize=(10, 6))
top_vendedoresxmonto.venta_pen.head(10).plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('Top 10 Vendedores por monto de ventas', fontsize=16)
plt.xlabel('Vendedor', fontsize=12)
plt.ylabel('Monto en PEN', fontsize=12)


plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
data['fecha'] = pd.to_datetime(data['fecha'], errors='coerce')

data['mes'] = data['fecha'].dt.to_period('M')

ventas_por_mes = data.groupby('mes')[['venta_usd', 'venta_pen']].sum()

ventas_por_mes
```

```python
top_10_meses = ventas_por_mes.sort_values(by='venta_pen', ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_10_meses['venta_pen'].plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Top 10 Meses con Más Ventas (PEN)', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas en PEN', fontsize=12)

# Mostrar el gráfico
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
